```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
xpoints=np.array([0,6])
ypoints=np.array([0,250])
```

```python
plt.plot(xpoints,ypoints)
plt.show()            #### If we give one datapoints like xpoints or y points it is taking by default to y axis and x axis is creating automatically. 
```

```python
#we can also increase the width of line
```

```python
plt.plot(xpoints,ypoints,lw='10')  # lw = line width
plt.show() 
```

```python
## Marker-- marker is used to highlight where the x axis point and y axis point is coordinating.
```

```python
plt.plot(xpoints,ypoints,marker='*')
plt.show()
```

```python
##format marker is written with this syntax:

## marker|line|color
```

```python
plt.plot(xpoints,ypoints,'*-.r')     ##     '-'	Solid line	
                                     ##     ':'	Dotted line	
                                     ##     '--'	Dashed line	
                                     ##     '-.'	Dashed/dotted line

plt.show()
```

```python
## we have different styles for marker i.e marker size as ms and markeredge colour as mec and markerface colour as mfc.
```

```python
plt.plot(xpoints,ypoints,'o:g',ms=30,mec='b',mfc='hotpink')
plt.show()
```

```python
## Labels
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
plt.plot(x1,y1)
plt.xlabel('Xaxis',size='20',color='r')
plt.ylabel('Yaxis',size='20',color='g')
plt.title('Naveen',size='20',color='darkred')  #if you want tittle in the left or right then use loc='left'
plt.show()
```

```python
## for grid lines in the graph use grid func
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
plt.plot(x1,y1)
plt.xlabel('Xaxis',size='20')
plt.ylabel('Yaxis',size='20')
plt.title('Naveen',size='20')  #if you want tittle in the left or right then use loc='left'
plt.grid()
plt.show()
```

```python
# we can also make grid line colourful
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
plt.plot(x1,y1)
plt.xlabel('Xaxis',size='20')
plt.ylabel('Yaxis',size='20')
plt.title('Naveen',size='20')  #if you want tittle in the left or right then use loc='left'
plt.grid(color='r',ls=':',lw=0.5)
plt.show()
```

```python
## subplot func if we want two different graphs side by side then we use subplot func
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1,2,1)   # here 1,2,1 is 1= rows, 2= coloumns, 1= 1st plot
plt.plot(x,y)
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.suptitle('Naveen')  ## super title alias suptitle is for main title
plt.show()
```

```python
## pie chart
```

```python
x = np.array([5,10,15,20])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.pie(x)
plt.show()
```

```python
## we can make changes in pie chart
```

```python
mylabels=['Naveen','Charan','Ravi','Nava']
plt.pie(x,labels=mylabels,startangle=180)
plt.show()
```

```python
mylabels=['Naveen','Charan','Ravi','Nava']
myexplode = [0.2,0, 0, 0]
plt.pie(x,labels=mylabels,explode=myexplode,shadow=True)
plt.legend()
plt.show()
```
